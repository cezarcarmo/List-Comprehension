## Multiplicado itens da lista (FOR)

```python
lst = [1,2,3,4,5]
mult5 = []

for numero in lst:
    n = numero * 5
    mult5.append(n)
    
print(mult5)
```

## List Comprehension

```python
mult5_lc = [numero * 5 for numero in lst]; print(mult5_lc)
```

## Tabuada com List Comprehension

```python
def tabuada(n):
    lst = range(1,11)
    resultado = [numero * n for numero in lst]
    return (f'Resultados da Tabuada do {n}: {resultado}')

tabuada(7)
```

## List Comprehension + Condicional

```python
## Sintaxes Possveis

# 1) [expresso for cada elemento in lista if condio] if

# 2) [expresso if condio else expresso for elemento in lista] if else
```

```python
lst = [100,45,294,23904,395,3864,87495,233,2245,67,788,998,776,900]

# sintaxe 1

[x for x in lst if x < 300]
```

```python
# sintaxe 2

[x if x < 300 else "Maior" for x in lst]
```

## Dictionary Comprehension

```python
## Sintaxes Possveis

# 1) {key: value for elemento in lista}

# 2) {key: value for (key, value) in dictionary.items()}
```

```python
frutas = ['banana', 'ma', 'pera', 'tomate', 'uva'] # Lista
```

```python
{i:frutas[i] for i in range(len(frutas))}
```

```python
d = {'a': 1, 'b':2, 'c':3} # Dicionrio
```

```python
{k:v*2 for (k,v) in d.items()} # Dobra o valor do Dicionrio
```
